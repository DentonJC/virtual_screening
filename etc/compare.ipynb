```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

def read_datafile(file_name):
    data = np.loadtxt(file_name, delimiter=';', skiprows=1)
    return data

data_reg_hiv = read_datafile('history/history_regression_HIV_maccs.csv')
x_reg_hiv = data_reg_hiv[:,0]
y_reg_hiv = data_reg_hiv[:,3]

data_res_hiv = read_datafile('history/history_residual_HIV_maccs.csv')
x_res_hiv = data_res_hiv[:,0]
y_res_hiv = data_res_hiv[:,3]

data_reg_muv = read_datafile('history/history_regression_muv_maccs.csv')
x_reg_muv = data_reg_muv[:,0]
y_reg_muv = data_reg_muv[:,3]

data_res_muv = read_datafile('history/history_residual_muv_maccs.csv')
x_res_muv = data_res_muv[:,0]
y_res_muv = data_res_muv[:,3]

data_reg_tox = read_datafile('history/history_regression_tox21_maccs.csv')
x_reg_tox = data_reg_tox[:,0]
y_reg_tox = data_reg_tox[:,3]

data_res_tox = read_datafile('history/history_residual_tox21_maccs.csv')
x_res_tox = data_res_tox[:,0]
y_res_tox = data_res_tox[:,3]


fig = plt.figure()
fig.set_size_inches(18.5, 10.5)

ax1 = fig.add_subplot(221)

ax1.set_title("HIV")    
ax1.set_xlabel('epochs')
ax1.set_ylabel('val_acc')

ax1.plot(x_reg_hiv, y_reg_hiv, c='r', label='regression')
ax1.plot(x_res_hiv, y_res_hiv, c='g', label='residual')

leg = ax1.legend()


ax2 = fig.add_subplot(222)

ax2.set_title("muv")    
ax2.set_xlabel('epochs')
ax2.set_ylabel('val_acc')

ax2.plot(x_reg_muv, y_reg_muv, c='r', label='regression')
ax2.plot(x_res_muv, y_res_muv, c='g', label='residual')

leg = ax2.legend()


ax3 = fig.add_subplot(223)

ax3.set_title("tox21")    
ax3.set_xlabel('epochs')
ax3.set_ylabel('val_acc')

ax3.plot(x_reg_tox, y_reg_tox, c='r', label='regression')
ax3.plot(x_res_tox, y_res_tox, c='g', label='residual')

leg = ax3.legend()

fig.savefig('val_acc.png', dpi=100)
plt.show()
```
